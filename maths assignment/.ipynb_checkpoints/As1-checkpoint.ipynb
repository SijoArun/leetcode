```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic dataset with 1500 rows and 15 columns
np.random.seed(0)
num_samples = 1500
num_features = 15
X = np.random.rand(num_samples, num_features)
y = np.random.rand(num_samples, 1)

# Initialize theta and set hyperparameters
theta_initial = np.zeros((num_features, 1))
learning_rate = 0.01
num_iterations = 1000

# Define your cost function and its gradient here
def cost_function(theta, X, y):
    predictions = X.dot(theta)
    errors = predictions - y
    cost = (1 / (2 * len(y))) * np.sum(errors ** 2)
    return cost

def gradient(theta, X, y):
    gradients = (1 / len(y)) * X.T.dot(X.dot(theta) - y)
    return gradients

def vanilla_gradient_descent(X, y, theta, learning_rate, num_iterations):
    cost_history = []

    for _ in range(num_iterations):
        cost = cost_function(theta, X, y)
        gradient_val = gradient(theta, X, y)
        theta = theta - learning_rate * gradient_val
        cost_history.append(cost)

    return theta, cost_history

def mini_batch_gradient_descent(X, y, theta, learning_rate, num_iterations, batch_size):
    m = len(y)
    cost_history = []

    for _ in range(num_iterations):
        indices = np.random.choice(m, batch_size, replace=False)
        X_batch = X[indices]
        y_batch = y[indices]
        cost = cost_function(theta, X_batch, y_batch)
        gradient_val = gradient(theta, X_batch, y_batch)
        theta = theta - learning_rate * gradient_val
        cost_history.append(cost)

    return theta, cost_history

def stochastic_gradient_descent(X, y, theta, learning_rate, num_iterations):
    m = len(y)
    cost_history = []

    for _ in range(num_iterations):
        for i in range(m):
            random_index = np.random.randint(0, m)
            X_i = X[random_index:random_index+1]
            y_i = y[random_index:random_index+1]
            cost = cost_function(theta, X_i, y_i)
            gradient_val = gradient(theta, X_i, y_i)
            theta = theta - learning_rate * gradient_val
            cost_history.append(cost)

    return theta, cost_history

# Experiment on Convergence for Mini-Batch Gradient Descent with different batch sizes
batch_sizes = [16, 32, 64, 128, 256]  # Experiment with different batch sizes
convergence_data = {}

for batch_size in batch_sizes:
    theta_initial = np.zeros((num_features, 1))
    _, cost_history_mini_batch = mini_batch_gradient_descent(X, y, theta_initial, learning_rate, num_iterations, batch_size)
    convergence_data[batch_size] = cost_history_mini_batch

# Plot cost history for Mini-Batch Gradient Descent with different batch sizes
plt.figure(figsize=(12, 6))
for batch_size, cost_history in convergence_data.items():
    plt.plot(range(num_iterations), cost_history, label=f"Batch Size: {batch_size}")

plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.legend()
plt.title("Convergence Comparison for Mini-Batch Gradient Descent")
plt.show()

```
